```python
import requests
from bs4 import BeautifulSoup
```

```python
#http규격에 맞춰 요청과 응답을 처리하는 모듈
response = requests.get('https://movie.daum.net/moviedb/grade?movieId=129982')
response.content
```

```python
#html코드로 parser
soup = BeautifulSoup(response.content, 'html.parser')
#class="ratings rating_10"
table = soup.find('div', {'class': 'cmt_info'})
print(table)
```
